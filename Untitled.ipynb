```python

MAX=201
STEP=1
r = range(0,MAX,STEP)
def f(x,y,z):
	if x ==0 and y==0 and z==0:
		return -1
	if (x+y+z-x*y*z-2)<=0 and (x==y and y==z):
		return -1
	return (x + y + z - x*y*z - 2)/(x*x + y*y + z*z - x*y - y*z - z*x)
max = -1
xr=yr=zr=0
for x in r:
	for y in r:
		for z in r:
			result =f(x/100.0,y/100.0,z/100.0) 
			if (result >= max):
				max = result
				xr=x/100.0
				yr=y/100.0
				xr=x/100.0
print(max)
print(xr,yr,zr)

```
